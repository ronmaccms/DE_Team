```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
from pygam import LinearGAM, s

# Assume merged_df is already loaded
# merged_df = pd.read_csv('your_data.csv')  # Example loading of data

# Step 2: Define a function to impute missing values using pyGAM for numeric columns only
def impute_with_pygam(df, target_column):
    # Ensure the target column is numeric
    if not np.issubdtype(df[target_column].dtype, np.number):
        raise ValueError(f"Target column '{target_column}' must be numeric.")

    # Split the data into features (X) and target (y)
    X = df.drop(columns=[target_column])
    y = df[target_column]

    # Filter out non-numeric columns from the features
    X = X.select_dtypes(include=[np.number])

    # Identify rows with missing target values
    missing_mask = y.isna()

    # Separate the data into training and prediction sets
    X_train = X[~missing_mask]
    y_train = y[~missing_mask]
    X_missing = X[missing_mask]

    # Impute missing values in the features using the mean of each column
    X_train = X_train.fillna(X_train.mean())
    X_missing = X_missing.fillna(X_train.mean())

    # Display the separated data (optional)
    print("Training Features (X_train):\n", X_train)
    print("Training Target (y_train):\n", y_train)
    print("Missing Features (X_missing):\n", X_missing)

    # Fit the pyGAM model
    gam = LinearGAM().fit(X_train, y_train)

    # Predict the missing values
    y_missing_pred = gam.predict(X_missing)

    # Fill the missing values in the original DataFrame
    df.loc[missing_mask, target_column] = y_missing_pred

    return df

# Step 3: Apply imputation to the 'Cost of Living Index' column
target_column = 'Cost of Living Index'
if merged_df[target_column].isna().sum() > 0:
    merged_df2 = impute_with_pygam(merged_df, target_column)

# Display the DataFrame after imputation
merged_df2.head()


```

### FIlling the NaN's using averages

```python
# for numeric data only, replace the NaN's with the average values of the same columns who share a similar Country, change the data frame name to merged_filled1_ df, for numeric data only round the floats to one decimal number

# Create a copy of the dataframe
merged_filled1_df = merged_df.copy()

# Select numeric columns
numeric_columns = merged_filled1_df.select_dtypes(include=['float64', 'int64']).columns

# Group the dataframe by country and calculate the mean for each numeric column
country_means = merged_filled1_df.groupby('Country')[numeric_columns].mean().round(1)

# Replace NaN values in the numeric columns with the corresponding country mean
for col in numeric_columns:
    for country in country_means.index:
        merged_filled1_df.loc[(merged_filled1_df['Country'] == country) & (merged_filled1_df[col].isna()), col] = country_means.loc[country, col]

# Display the first few rows of the dataframe
merged_filled1_df.head()
merged_filled1_df.shape
```
