```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from bs4 import BeautifulSoup
print(selenium.__version__)
```

```python

# Initialize WebDriver
driver = webdriver.Chrome()  # Assuming you're using Chrome

# List of country codes
country_codes = ['DZA', 'MAR', 'NLD', 'ESP', 'VNM', 'DNK', 'DEU', 'CHE', 'LBN', 'BEL', 'FIN', 'IRN',
                 'CHN', 'NGA', 'BRA', 'RUS', 'PER', 'CRI', 'ITA', 'JPN', 'LKA', 'MEX', 'KAZ', 'FRA',
                 'USA', 'IND', 'GBR', 'CAN', 'PAK', 'COL', 'SAU', 'IRQ', 'ETH', 'EGY', 'CMR', 'HUN',
                 'PHL', 'GHA', 'ARM', 'ARG', 'PRT', 'ARE', 'KWT', 'VEN', 'TUR', 'GTM', 'GRC', 'ZAF',
                 'BHR', 'SRB', 'LVA', 'AUS', 'QAT', 'ROU', 'IDN', 'ISR', 'TUN', 'AZE', 'CYP', 'PRI',
                 'PAN', 'SWE', 'URY', 'SVN', 'KGZ', 'THA', 'GEO', 'ISL', 'OMN', 'JOR', 'ECU', 'UKR',
                 'POL', 'CHL', 'NOR', 'MYS', 'DOM', 'MNG', 'AUT', 'LTU', 'NPL', 'BGD', 'IRL', 'UZB',
                 'MNE', 'MDA', 'MKD', 'BGR', 'BLR', 'NZL', 'KEN', 'MLT', 'TWN', 'ALB', 'KHM', 'SVK',
                 'ZWE', 'HRV', 'RWA', 'LUX', 'EST', 'BMU', 'SGP']



# Base URL
url_base = 'https://databank.worldbank.org/views/reports/reportwidget.aspx?Report_Name=CountryProfile&Id=b450fd57&tbar=y&dd=y&inf=n&zm=n&country='

# Initialize an empty DataFrame
df_countries = pd.DataFrame(columns=["key", "1990", "2000", "2010", "2021", "country_code"])

# Loop through each country code
for country_code in country_codes:
    # Construct URL for each country
    url_string = url_base + country_code
    print("Fetching data for:", country_code)
    print("URL:", url_string)
    
    # Fetch data using Selenium
    driver.get(url_string)
    
    # Find Table element
    tbl = driver.find_element(By.ID, 'container-data')  # Assuming the table has this ID
    
    # Parse Table element into pandas DataFrame
    table_html = tbl.get_attribute('outerHTML')
    soup = BeautifulSoup(table_html, 'html.parser')
    
    # Find all table rows in the parsed HTML
    table_rows = soup.find_all('tr')
    
    # Initialize data list
    data = []
    for row in table_rows:
        row_data = []
        cells = row.find_all(['td', 'th'])
        for cell in cells:
            row_data.append(cell.get_text().strip())
        data.append(row_data)

    # Print out data structure for debugging
    print("Columns:", data[0])
    print("Data:", data[1:])

    # Create DataFrame & append to existing one
    df_aux = pd.DataFrame(data[2:], columns=df_countries.columns[0:5])  # Assuming first row is header
    df_aux['country_code'] = country_code
    df_countries = pd.concat([df_countries, df_aux], ignore_index=True)




# Close WebDriver
driver.quit()

# Display the final DataFrame
print(df_countries)

```
