```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from bs4 import BeautifulSoup
print(selenium.__version__)
```

```python
# Initialize WebDriver
driver = webdriver.Chrome()  # Assuming you're using Chrome

```

```python
url_liv_cost ='https://www.numbeo.com/cost-of-living/rankings.jsp'
```

```python
df_capital = pd.DataFrame()
```

```python
# Fetch data using Selenium
driver.get(url_liv_cost)

# Find Table element
tbl_cost = driver.find_element(By.XPATH, '//*[@id="t2_wrapper"]')  # if there's no ID, always a good idea to copy xpath

# Parse Table element into pandas DataFrame
table_cost_html = tbl_cost.get_attribute('outerHTML')
soup = BeautifulSoup(table_cost_html, 'html.parser')

# Find all table rows in the parsed HTML
table_cost_rows = soup.find_all('tr')
```

```python
# Initialize data list
data_cost = []
for row in table_cost_rows:
        row_data = []
        cells = row.find_all(['td', 'th'])
        for cell in cells:
            row_data.append(cell.get_text().strip())
        data_cost.append(row_data)

# Print out data structure for debugging
print("Columns:", data_cost[0])
print("Data:", data_cost[1:])
```

```python
# Extract header and data rows separately
header = data_cost[0]
data = data_cost[1:]

# Create DataFrame
df_capital = pd.DataFrame(data, columns=header)

# Display DataFrame
print(df_capital)
df_capital.head(20)
```
